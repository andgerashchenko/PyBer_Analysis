### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
city_type_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
city_type_ride_count.name = 'Total Rides'
print(city_type_ride_count)
```

```python
# 2. Get the total drivers for each city type
city_type_drivers_sum = city_data_df.groupby(["type"]).sum()["driver_count"]
city_type_drivers_sum.name = 'Total Drivers'
city_type_drivers_sum.head(20)
```

```python
#  3. Get the total amount of fares for each city type
city_type_fares_sum = pyber_data_df.groupby(["type"]).sum()["fare"]
city_type_fares_sum.name = 'Total Fares'
city_type_fares_sum.head(20)
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = pyber_data_df.groupby(["type"]).sum()["fare"] / pyber_data_df.groupby(["type"]).count()["ride_id"]
average_fare_per_ride.name = 'Averaga Fare per Ride'
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = city_type_fares_sum / city_type_drivers_sum
average_fare_per_driver.name = 'Agerage Fare per Driver'
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. rage_f
from functools import reduce
df = [city_type_ride_count,  city_type_drivers_sum, city_type_fares_sum, average_fare_per_ride, average_fare_per_driver]
pyber_summary_df = reduce(lambda  left,right: pd.merge(left,right,on=['type'], how='outer'), df)

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Averaga Fare per Ride"] = pyber_summary_df["Averaga Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Agerage Fare per Driver"] = pyber_summary_df["Agerage Fare per Driver"].map("${:.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
type_fare_sum = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
type_fare_sum
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
type_fare_sum = type_fare_sum.reset_index()
type_fare_sum
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
type_fare_sum_pivot = type_fare_sum.pivot(index="date", columns="type", values="fare")
type_fare_sum_pivot.reset_index(inplace=True)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
type_fare_sum_april = type_fare_sum_pivot.loc[type_fare_sum_pivot["date"] <= "2019-04-29", :]
type_fare_sum_april = type_fare_sum_april.set_index('date')
type_fare_sum_april
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
type_fare_sum_april.index = pd.to_datetime(type_fare_sum_april.index)
type_fare_sum_april
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
type_fare_sum_april.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
type_fare_sum_april = type_fare_sum_april.resample("W").sum()
type_fare_sum_april                                             
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
plt.style.use('fivethirtyeight')
type_fare_sum_april.plot(figsize = (12, 4))
plt.title("Total Fare by City Type")
plt.legend(loc = 'center')
plt.ylabel("Fare($USD)")
plt.xlabel("")
plt.savefig("analysis/PyBer_fare_summary.png")
```
